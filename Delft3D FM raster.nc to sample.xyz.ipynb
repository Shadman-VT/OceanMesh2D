```python
import netCDF4 as nc
import numpy as np

def netcdf_to_xyz(netcdf_file, xyz_file, x_var, y_var, z_var):
    """
    Convert a NetCDF file to an XYZ format file and display progress.

    :param netcdf_file: Path to the NetCDF file.
    :param xyz_file: Path to the output XYZ file.
    :param x_var: Name of the variable representing the longitude in the NetCDF file.
    :param y_var: Name of the variable representing the latitude in the NetCDF file.
    :param z_var: Name of the variable representing the elevation in the NetCDF file.
    """
    with nc.Dataset(netcdf_file, 'r') as ds:
        x_data = ds.variables[x_var][:]
        y_data = ds.variables[y_var][:]
        z_data = ds.variables[z_var][:]

        total_iterations = len(y_data) * len(x_data)
        iteration = 0

        with open(xyz_file, 'w') as f:
            for i in range(len(y_data)):
                for j in range(len(x_data)):
                    f.write(f"{x_data[j]:.12f} {y_data[i]:.12f} {z_data[i, j]:.3f}\n")
                    iteration += 1
                    if iteration % 1000000 == 0:  # Update progress every 10000 iterations
                        progress = (iteration / total_iterations) * 100
                        print(f"Progress: {progress:.2f}%")

# Example usage
netcdf_file = r"E:\VirginiaTech-Research\SanguModel\Data\DEM\Batthy_NTT.nc"
xyz_file = r"C:\Users\Shadman\Downloads\SanfuNTT_DEM.xyz"

# Variables in your NetCDF file
x_var = 'lon'  
y_var = 'lat'
z_var = 'NTT_bathy.tif' ### This value has to be put in from the second part of the code

netcdf_to_xyz(netcdf_file, xyz_file, x_var, y_var, z_var)
```

```python
import netCDF4 as nc
import numpy as np

def netcdf_to_xyz(netcdf_file, xyz_file, x_var, y_var, z_var):
    """
    Convert a NetCDF file to an XYZ format file and display progress.
    Skip writing lines where z_var data is missing or '--'.

    :param netcdf_file: Path to the NetCDF file.
    :param xyz_file: Path to the output XYZ file.
    :param x_var: Name of the variable representing the longitude in the NetCDF file.
    :param y_var: Name of the variable representing the latitude in the NetCDF file.
    :param z_var: Name of the variable representing the elevation in the NetCDF file.
    """
    with nc.Dataset(netcdf_file, 'r') as ds:
        x_data = ds.variables[x_var][:]
        y_data = ds.variables[y_var][:]
        z_data = ds.variables[z_var][:]

        total_iterations = len(y_data) * len(x_data)
        iteration = 0

        with open(xyz_file, 'w') as f:
            for i in range(len(y_data)):
                for j in range(len(x_data)):
                    z_value = -z_data[i, j]
                    # Check if z_value is missing or '--'
                    if not np.isnan(z_value) and z_value != '--':
                        f.write(f"{x_data[j]} {y_data[i]} {z_value}\n")
                    iteration += 1
                    if iteration % 100000 == 0:  # Update progress every 1,000,000 iterations
                        progress = (iteration / total_iterations) * 100
                        print(f"Progress: {progress:.2f}%")

# Example usage
netcdf_file = r"D:\OneDrive - Virginia Tech\03_Research\Hydrodynamic\Eastern Hill River\Main DEM\DEM_Sangu.nc"
xyz_file = r'D:\OneDrive - Virginia Tech\03_Research\Hydrodynamic\Eastern Hill River\Main DEM\Sangu_FABDEM_Version4.xyz'

# Variables in your NetCDF file
x_var = 'lon'  
y_var = 'lat'
z_var = 'DEM_Clip_Filled'

netcdf_to_xyz(netcdf_file, xyz_file, x_var, y_var, z_var)
```
